```
def label_one_five(id:int)->bool:
  if id == 1:
    return 1
  #elif id ==5:
    #return 0
  elif math.isnan(id):
    return np.nan
  else:
    return 0

def label_one_two(id:int)->bool:
  if id == 1:
    return 1
  elif id ==2: 
    return 0
  elif math.isnan(id):
    return np.nan
  else:
    return -1


def one_five_func(df, cols):
  for col in cols:
    df[col] = df[col].apply(label_one_five)

def one_two_func(df, cols):
  for col in cols:
    df[col] = df[col].apply(label_one_five)

def merge(df, cols, name, inverse=False):

  if inverse == True:
    if len(cols) != 2:
      return -1

    a = df[cols[0]]
    b = np.invert(df[cols[1]])
    df[name] = np.logical_or(a, b)

  else:
    
    df[name] = df[cols[0]]
  #pd.Series([np.nan for x in range(len(df.index))], index=df.index)

    for col in df[cols].columns:
      df[name] = np.logical_or(df[name], df[col])

    df[name] = df[name]*1

```

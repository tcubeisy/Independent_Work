```
races_data = {name: data[data[col] == 1] for name, col in races_columns}
```

# Races Association Rules

## White

```
white_data = races_data["white"]
white_frequent = apriori(df=white_data, min_support=0.35, use_colnames=True)
white_frequent
```

```
white_rules = association_rules(df=white_frequent, metric="lift", min_threshold=1.2)
white_rules.sort_values(by="lift", ascending=False)
```

## Black

```
black_data = races_data["black"]
black_frequent = apriori(df=black_data, min_support=0.35, use_colnames=True)
black_frequent
```

```
black_rules = association_rules(df=black_frequent, metric="lift", min_threshold=1.2)
black_rules.sort_values(by="lift", ascending=False)
```

```
hisp_rules = association_rules(df=hisp_frequent, metric="lift", min_threshold=1.2)
hisp_rules.sort_values(by="lift", ascending=False)
```
